# El DataFrame de Pandas sirve para trabajar con estructuras de dos dimensiones y se pueden cargar de 2 maneras, manualmente o leyendo de un fichero, base de datos o cualquier origen de datos.

    * Cargando un DataFrame de manera manual

```python
import pandas as pd 
clase = pd.DataFrame([{"Nombre":"Pedro", "Estatura":1.71,"Edad":28},
                      {"Nombre":"Maria", "Estatura":1.73,"Edad":45},
                      {"Nombre":"Andrea", "Estatura":1.68,"Edad":61},
                      {"Nombre":"Marcos", "Estatura":1.65,"Edad":19},
                      ])
clase
```

```python
# Como ver solamente un campo o columna
clase["Nombre"]
```

```python
clase["Edad"].sort_values(ascending=True)
```

```python
# FUNCIONES DE RESUMEN
print(clase["Edad"].sum())
print(clase["Estatura"].max())
print(clase["Estatura"].min())
print(f"El promedio de edad es {(round(clase["Edad"].mean(),2))} años")
```

# COMO AÑADIR COLUMNAS A UN DF.
* OPCION A: PONIENDO UN VALOR MANUAL FIJO
* OPCION B: HACIENDO UN CÁLCULO A PARTIR DE UNA O VERIAS COLUMNAS EXISTENTES
* OPCION C: APLICANO UNA FUNCIÓN 


```python
# OPCION A:
clase["Colegio"]= "IT Academy"
clase
```

```python
# OPCION B:
clase["Edad2"]= clase["Edad"]+2
clase
```

```python
clase["Diferencia"] = clase["Edad2"]-clase["Edad"]
clase
```

```python
# Incrementas la estartura de cada alumno en un 10%
clase["Estatura2"]=clase["Estatura"]+clase["Estatura"]*0.1
clase["Estatura3"]=clase["Estatura"]*1.1    # manera de hacerlo directo
clase
```

```python
# OPCION C: con apply, restar 2 años a los alumnos que tengas mas de 30 años
def resta2(edad):
    if edad > 30:
        return edad-2
    else:
        return edad
    
# Crear la columna nueva con apply
clase["Edad5"]= clase["Edad"].apply(resta2)
clase
```

```python
# COMO ELIMINAR UNA COLUMNA CON LA INSTRUCCION DROP + EL PARÁMETRO AXIS
# AXIS 1 SIGNIFICA COLUMNAS Y AXIS 0 FILAS
clase=clase.drop(["Estatura3"], axis = 1)
clase
```

    * Esta es la manera PRO

```python
# Elimina directo sin tener que asignarlo a el mismo
clase.drop(["Edad2"], axis = 1, inplace=True)
clase
```

```python
# Para eliminar una fila con DROP pero con AXIs = 0
clase.drop([2], axis=0,inplace=True)
clase
```
