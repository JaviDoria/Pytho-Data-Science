EN UN FICHERO NUEVO DE PYTHO:
CARGAREMOS EL FICHERO TITANIC CSV Y RESOLVEREMOS LOS SIGUIENTES ENUNCIADOS.


```python
import pandas as pd 
titanic = pd.read_csv(r"C:\Users\javie\Downloads\Titanic.csv")  
titanic
```

```python
# VER LOS 6 PRIMEROS NOMBRES DE LOS PASAJEROS DE LA CLASE 1 EN ORDEN DESCENDENTE
titanicC1=titanic["Pclass"]==1
titanic[titanicC1]["Name"].head(6).sort_values(ascending=False)
```

```python
# VER LAS 3 ULTIMAS EDADES DE LOS PASAJEROS DE LA CLASE 3
titanicC3=titanic["Pclass"]==3
titanic[titanicC3]["Age"].tail(3)
```

```python
# VER LOS 4 PRIMEROS NOMBRES DE LOS PASAJEROS QUE NO SEAN DE LA CLASE 2
titanicC2=titanic["Pclass"]!=2
titanic[titanicC2]["Name"].head(4)
```

```python
# VER LOS 6 PRIMEROS PASAJEROS CUYO NOMBRE CONTIENE LA PALABRA MARY
titanicMary=titanic["Name"].str.contains("Mary")
titanic[titanicMary].head(6)
```

```python
# VER LOS 10 PRIMEROS NOMBRES DE LOS PASAJEROS QUE SOBREVIVIERON (SURVIVED) 
titanicSurv=titanic["Survived"]==1
titanic[titanicSurv]["Name"].head(10)
```

```python
# Como saber el numero de perosnas de la clase 1
numero=titanic[titanicC1].PassengerId.count()
print(f"El número de pasajeros de la clase 1 es de {numero}")
```

```python
# Como aplicar más de una condicion en el filtro, para ello usaremos los operadores lógicos OR + and.
# En el caso del OR untilizaremos este carácter | 
# En el caso de AND usaremos &
titanicC13 = (titanic["Pclass"]==1)| (titanic["Pclass"]==3) # Parentesis en cada condicion
titanic[titanicC13].PassengerId.count()
```

```python
# Pasajeros de la clase 2 que embarcaron por la puerta s
titanicC2Ps = (titanic["Pclass"]==2)&(titanic["Embarked"]=="S")
titanic[titanicC2Ps]
```

```python
# ver los diferentes valores de un campo usaremos unique
titanic["Pclass"].unique()
```

```python
# Práctica 10: Ver la cantidad de pasajeros que son hombres de la clase 1 y que embarcaron por la puerta s
titanicC1HS=(titanic["Sex"]=="male") & (titanic["Embarked"]=="S") & (titanic["Pclass"]==1)
titanic[titanicC1HS].PassengerId.count()
```

```python
# Práctica 11: ver los nombres de los pasajeros de la clase 3 que no sobrevivieron
titanicNC3NS=(titanic["Survived"]==0)&(titanic["Pclass"]==3)
titanic[titanicNC3NS].Name
```

```python
# Práctica 12: Ver los 12 ultimos nombres en orden descendente de los pasajeros que son mujeres de la clase 3 y cuyo Passenger id es mayor a 100
titanic12=(titanic["Sex"]=="female")&(titanic["Pclass"]==3)&(titanic["PassengerId"]>100)
titanic[titanic12].Name.tail(12).sort_values(ascending=False)
```

```python
# Práctica 13: ver una lista con los nombres cuyo passenger id esté comprendido entre 70 y 180
titanic70100=(titanic["PassengerId"]>= 70) & (titanic["PassengerId"]<= 180)
titanic[titanic70100].Name
```
