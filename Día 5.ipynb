    * Partiendo del fichero ventas grocery meses en formato de excel , cargar los 3 dataframes de los meses de agosto a octubre del 2014 .

```python
import pandas as pd
fichero = r"C:\Users\javie\Downloads\VentasGroucery_Meses.xlsx"
           
agosto2014=pd.read_excel(fichero, sheet_name="Ventas Agosto 2014")   

septiembre2014=pd.read_excel(fichero, sheet_name="Ventas Septiembre 2014") 

octubre2014=pd.read_excel(fichero, sheet_name="Ventas Octubre 2014") 
```

    * Realizar las modificaciones necesarias, despues de anexar todos los dataframes en uno solo al que llamaremos trimestre y resolveremos los siguientes enunciados:

```python
trimestre = pd.concat([agosto2014,septiembre2014,octubre2014], ignore_index=True) 
trimestre
```

    * Crear la columna amount que será igual a unit price * quantity

```python
trimestre["Amount"]=trimestre["Unit Price"]*trimestre["Quantity"]
trimestre
```

   * Crear la columna enterprise que será igual al " Delta Enterprise"

```python
trimestre["Enterprise"]="Delta Enterprise"
trimestre
```

    * Crear la columna discount = 3.35% sobre el amount , solo para los pedidos que sean de France o Germany.

```python
def descuento(pais):
    if pais == "France" or pais == "Germany":
        return 0.0335
    else:
        return 0 

trimestre["Discount"] = trimestre["Amount"] * trimestre["Ship Country"].apply(descuento)

trimestre
```

    * Ver en un texto la suna de discount ( "El total de descuento es x ")

```python
total_Discount = trimestre["Discount"].sum()
print(f"El total del descuento es: {round(total_Discount,2)}")
```

    * Ver en un texto el promedio de amount, de los pedidos cuya categoría = Beverages. ("El promedio es X")

```python
catbeverages = trimestre["Category"]=="Beverages"
media_catbeverages = trimestre[catbeverages]["Amount"].mean()
print(f"La media de los amount de la categoría Beverages es {round(media_catbeverages,2)}")
```

    * Ver en un texto la suma de amount de los pedidos de la categoria seafood cuya 
    cantidad esté comprendido entre 25 y 75.("La suma de amount de seafood es x ")

```python
amount_catseafood = (trimestre["Quantity"] >=25) & (trimestre["Quantity"] <= 75) & (trimestre["Category"] == "Seafood")

suma_amount = trimestre[amount_catseafood]["Amount"].sum()

print(f"La suma del amount de los pedidos de la categoría Seafood cuya cantidad está comprendida entre 25 y 75 es {round(suma_amount, 2)}")
```
